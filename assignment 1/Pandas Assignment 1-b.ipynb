#### #6. Sort the data set in on the `players` column in alphabetical order.

```python
nba_df = nba_df.sort_values(by='player', ascending=True)
nba_df.reset_index(drop=True, inplace=True)
nba_df.head()
```

#### #7. Filter the data set. Create three sub DataFrames from the `position` column for `G`, `F`, and `C`.

```python
nba_df_G = nba_df[nba_df['pos'] == 'G']
nba_df_G.reset_index(drop=True, inplace=True)
nba_df_G.head()
```

```python
nba_df_F = nba_df[nba_df['pos'] == 'F']
nba_df_F.reset_index(drop=True, inplace=True)
nba_df_F.head()
```

```python
nba_df_C = nba_df[nba_df['pos'] == 'C']
nba_df_C.reset_index(drop=True, inplace=True)
nba_df_C.head()
```

```python
print(nba_df_G.shape, nba_df_F.shape, nba_df_C.shape)
```

#### #8.  Run `describe()` on these new DataFrames. Compare the mean field goals (the `fg` column) between positions.

```python
print('G: ' ,nba_df_G['fg'].mean())
print('F: ' ,nba_df_F['fg'].mean())
print('C: ' ,nba_df_C['fg'].mean())
```
